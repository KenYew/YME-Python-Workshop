We can have an idea of how the maximum concentration of a pollutant varies over the year by plotting the max value for every month.

Let's do it for the NO concentration in the area of Greenwich.

```python
months=['01','02','03','04','05','06','07','08','09','10','11','12']
GN_NO_data_MaxValues=[]
for idx,month in enumerate(months):
    month_start_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='01/'+months[idx]+'/2018 00:00'].index[0]
    if idx==11:
        month_end_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='30/12/2018 00:00'].index[0]
    else:
        month_end_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='01/'+months[idx+1]+'/2018 00:00'].index[0]-1

    maxloc=GN_NO_data.loc[month_start_index:month_end_index]['Value'].idxmax()
    GN_NO_data_MaxValues.append(GN_NO_data.loc[maxloc]['Value'])
plt.subplot()
plt.plot(GN_NO_data_MaxValues)
plt.title('Monthly Max NO concentration in 2018 in area of GN')
plt.xticks(range(12),months)
plt.xlabel('month')
plt.ylabel(r'$\mu$ $g/m^{-3}$')
plt.show()
```

### 2.3.3 Mean value
Mean value is the average of a set of data.

For example, given the NO pollutant value in January is $a=[a_0,...,a_{31}]$, the mean value of NO concentration in January is 
\begin{equation}
E[a]=\frac{\sum_{i=0}^{31}a_{i}}{31}
\end{equation}


To evaluate the mean value of a given vector, we can use the Numpy function **mean()**.

**Q**: What is the average concentration of NO in 2018?

```python
NO_Data['Value'].mean()
```

**[TO DO]** Plot the monthly mean value of NO concentration in 2018 in the area of Greenwich.

```python
###########################
# Task: 
#   Plot the monthly mean value of NO concentration in 2018 in area of GN3.
###########################
months=['01','02','03','04','05','06','07','08','09','10','11','12']
GN_NO_data_MeanValues=[]
for idx,month in enumerate(months):
    month_start_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='01/'+months[idx]+'/2018 00:00'].index[0]
    if idx==11:
        month_end_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='30/12/2018 00:00'].index[0]
    else:
        month_end_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='01/'+months[idx+1]+'/2018 00:00'].index[0]-1

    mean_value=GN_NO_data.loc[month_start_index:month_end_index]['Value'].mean()
    GN_NO_data_MeanValues.append(mean_value)
plt.subplot()
plt.plot(GN_NO_data_MeanValues)
plt.title('Monthly Mean NO concentration in 2018 in area of GN')
plt.xticks(range(12),months)
plt.xlabel('month')
plt.ylabel(r'$\mu$ $g/m^{-3}$')
plt.show()

### to be completed
###########################

```

### 2.3.4 Variance
Variance is a metric that quantifies the deviation of a data set from its mean value.

Considering the example of the mean value, the variance can be defined as
\begin{equation}
Var[a]=E[(a-E[a])^2]=\frac{\sum_{i=0}^{30}(a_{i}-E[a])^2}{31}
\end{equation}

To evaluate the variance of a given vector, we can use the Numpy function **var()**.

**Q**: How does the concentration of NO vary in 2018?

```python
NO_Data['Value'].var()
```

**[TO DO]** Plot the monthly variance of NO concentration in 2018 in area of Greenwich.

```python
###########################
# Task: 
#   Plot the monthly variance value of NO concentration in 2018 for the area of GN3.
###########################
months=['01','02','03','04','05','06','07','08','09','10','11','12']
GN_NO_data_VarValues=[]

# for loop handles specification of the correct start and end dates for each month (e.g.: 1st and 28th for Feb but 1st and 31st for Jan)
for idx,month in enumerate(months):
    # Setting the start date for every month (all months share the same start date e.g.: 01/month)
    month_start_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='01/'+months[idx]+'/2018 00:00'].index[0]
    
    # Handling the end dates for every month (not all months share the same end date e.g.: 31/01 or 28/02 or 30/03)
    if idx==11:
        # Hard coding the end date of December as there is no 13th month to use as a reference to get the end date of December
        month_end_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='30/12/2018 00:00'].index[0]
    else:
        # The end date is determined by taking the start date of the next month - 1 
        month_end_index=GN_NO_data[GN_NO_data['ReadingDateTime']=='01/'+months[idx+1]+'/2018 00:00'].index[0]-1

    # Calculate the variance value for each month and append it to the array GN_NO_data_VarValues
    var_value=GN_NO_data.loc[month_start_index:month_end_index]['Value'].var()
    GN_NO_data_VarValues.append(var_value)
    
plt.subplot()
plt.plot(GN_NO_data_VarValues)
plt.title('Monthly Variance NO concentration in 2018 in area of GN')
plt.xticks(range(12),months)
plt.xlabel('month')
plt.ylabel(r'$\mu$ $g/m^{-3}$')
plt.show()

### to be completed
###########################
```

From the website:
"Particles or particulate matter (PM) are tiny bits of solids or liquids suspended in the air. They are a complex group of pollutants that vary in size, shape, composition and origin. These properties can also vary from place to place and time to time.

Particles originating from road traffic include carbon emissions from engines, small bits of metal and rubber from engine wear and braking as well as dust from road surfaces. Others include material from building and industry as well as wind-blown dust, sea salt, pollens and soil particles.

People often associate particles with smoky engines, but the majority of particles that can penetrate the airway are too small to see, and thus can be present in air that seems clean. 

Larger particles are generally filtered in the nose and throat and do not cause problems. Particles smaller than about 10 micrometers, referred to as PM10 and particles smaller that 2.5 micrometres (PM2.5), can settle in the airway and deep in the lungs and cause health problems.

The health effects of particle air pollution have been widely studied, and include premature death and the worsening of heart and lung disease, often increasing admissions to hosptial."
