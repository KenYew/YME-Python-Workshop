### 2.3.3 Mean value
Mean value is the average of a set of data.

For example, given the NO pollutant value in January is $a=[a_0,...,a_{31}]$, the mean value of NO concentration in January is 
\begin{equation}
E[a]=\frac{\sum_{i=0}^{31}a_{i}}{31}
\end{equation}


To evaluate the mean value of a given vector, we can use the Numpy function **mean()**.

**Q**: What is the average concentration of NO in 2018?

```python
NO_Data['Value'].mean()
```

**[TO DO]** Plot the monthly mean value of NO concentration in 2018 in the area of Greenwich.

### 2.3.4 Variance
Variance is a metric that quantifies the deviation of a data set from its mean value.

Considering the example of the mean value, the variance can be defined as
\begin{equation}
Var[a]=E[(a-E[a])^2]=\frac{\sum_{i=0}^{30}(a_{i}-E[a])^2}{31}
\end{equation}

To evaluate the variance of a given vector, we can use the Numpy function **var()**.

**Q**: How does the concentration of NO vary in 2018?

```python
NO_Data['Value'].var()
```

**[TO DO]** Plot the monthly variance of NO concentration in 2018 in area of Greenwich.

*We can either iterate in a single for loop over the 4 areas, or do it separately. The first generally results in a cleaner code.*

```python
#  [TO DO (3)] - CD9

CD_PM10_Data = PM10_Data[PM10_Data['Site'] == 'CD9'].set_index('ReadingDateTime')
GN_PM10_Data = PM10_Data[PM10_Data['Site'] == 'GN3'].set_index('ReadingDateTime')
BT_PM10_Data = PM10_Data[PM10_Data['Site'] == 'BT4'].set_index('ReadingDateTime')
KT_PM10_Data = PM10_Data[PM10_Data['Site'] == 'KT6'].set_index('ReadingDateTime')

CD_air_quality = []
GN_air_quality = []
BT_air_quality = []
KT_air_quality = []

for i in range(12):
    CD_air_quality.append(CD_PM10_Data.loc[dates[i][0]:dates[i][1], 'Value'].mean(skipna=True))
    GN_air_quality.append(GN_PM10_Data.loc[dates[i][0]:dates[i][1], 'Value'].mean(skipna=True))
    BT_air_quality.append(BT_PM10_Data.loc[dates[i][0]:dates[i][1], 'Value'].mean(skipna=True))
    KT_air_quality.append(KT_PM10_Data.loc[dates[i][0]:dates[i][1], 'Value'].mean(skipna=True))
    
print('For the area of Camden-Euston Road, the best air quality is in {}, with an average PM10 concentration of {:1.2f} ug/m-3'
      .format(months[np.nanargmin(CD_air_quality)], np.nanmin(CD_air_quality)))
print()
print('For the area of Greenwich, the best air quality is in {}, with an average PM10 concentration of {:1.2f} ug/m-3'
      .format(months[np.nanargmin(GN_air_quality)], np.nanmin(GN_air_quality)))
print()
print('For the area of Brent, the best air quality is in {}, with an average PM10 concentration of {:1.2f} ug/m-3'
      .format(months[np.nanargmin(BT_air_quality)], np.nanmin(BT_air_quality)))
print()
print('For the area of Kingston Upon Thames, the best air quality is in {}, with an average PM10 concentration of {:1.2f} ug/m-3'
      .format(months[np.nanargmin(KT_air_quality)], np.nanmin(KT_air_quality)))
```
